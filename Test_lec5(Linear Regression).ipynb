```python
import torch
import torch.nn as nn
import torch.optim
import numpy as np

x_data = torch.tensor([[1.0],[2.0],[3.0]])
y_data = torch.tensor([[2.0],[4.0],[6.0]])

class Model(nn.Module):
    
    def __init__(self):
        super(Model, self).__init__() #initialize the module
        self.linear = nn.Linear(1,1)

    def forward(self, x):
        y_pred = self.linear(x)
        return y_pred
model=Model()

crit = nn.MSELoss(size_average=False) #loss function
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) #optimizer

#training
for epoch in range(500):
    y_pred = model(x_data)
    
    loss = crit(y_pred, y_data)
    print(epoch, loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

#after
check = torch.tensor([4.0])
print("predict (after training)", 4,model.forward(check).item())
    
```
