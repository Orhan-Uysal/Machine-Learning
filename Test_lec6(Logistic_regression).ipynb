```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim

x_data = torch.tensor([[1.0],[2.0],[3.0],[4.0]])
y_data = torch.tensor([[0.],[0.],[1.0],[1.0]])

class Model(nn.Module):
    
    def __init__(self):
        super(Model,self).__init__()
        self.linear = nn.Linear(1,1)#one in one out
    
    def forward(self,x):
        y_pred = F.sigmoid(self.linear(x))
        return y_pred

model = Model()

crit = nn.BCELoss(size_average=False) #loss function
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) #optimizer

#training loop
for epoch in range(500):
    y_pred = model(x_data)
    
    loss = crit(y_pred, y_data)
    print(epoch, loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

check = torch.tensor([5.0])
print("predict (after training)", 1.0 ,model.forward(check).item() > 0.5)
```
