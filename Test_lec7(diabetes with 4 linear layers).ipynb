```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim
import numpy as np
import pandas as pd

xy = np.loadtxt('C:/Users/O/Desktop/dataset/diabetes.csv',delimiter=',',dtype=np.float32)

x_data = torch.from_numpy(xy[:,0:-1])
y_data = torch.from_numpy(xy[:,[-1]])

class Model(nn.Module):
    
    def __init__(self):
        super(Model,self).__init__()
        self.l1 = nn.Linear(8,6)#two in one out
        self.l2 = nn.Linear(6,4)#two in one out
        self.l3 = nn.Linear(4,3)#two in one out
        self.l4 = nn.Linear(3,1)
        
        self.sigmoid = nn.Sigmoid()
        
    def forward(self,x):
        out1 = self.sigmoid(self.l1(x))
        out2 = self.sigmoid(self.l2(out1))
        out3 = self.sigmoid(self.l3(out2))
        
        y_pred = self.sigmoid(self.l4(out3))
        return y_pred

model = Model()

crit = nn.BCELoss(size_average=False) #loss function
optimizer = torch.optim.SGD(model.parameters(), lr=0.01) #optimizer

#training loop
for epoch in range(750):
    y_pred = model(x_data)
    
    loss = crit(y_pred, y_data)
    print(epoch, loss.item())
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

#check = torch.tensor([[-0.764706],[0.0653266],[0.0491803],[-0.292929],[-0.718676],[-0.0909091],[0.12895],[-0.566667]])
#check = torch.tensor([[0.176471],[0.085427],[0.0819672],[0],[0],[-0.0342771],[-0.83433],[-0.3]])
#xyz = np.loadtxt('C:/Users/O/Desktop/dataset/dia2.csv',delimiter=',',dtype=np.float32)
#x_data2 = torch.from_numpy(xyz[:,0:-1])

cnt = 0
for i in range(750):
    check = x_data[i]
    check = check.view(-1,8)
    if(model.forward(check).item()<=0.5 and y_data[i] == 0):
        cnt+=1
    if(model.forward(check).item()>0.5 and y_data[i] == 1):
        cnt+=1

print(cnt)
```
